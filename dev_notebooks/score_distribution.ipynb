```python
def get_stableford_distribution_by_teg():
    """
    Returns a DataFrame showing the distribution of stableford points by TEG.
    
    Returns:
        pd.DataFrame: TEG as index, stableford points (0-max) as columns, 
                     cells show percentage of holes with that score
    """
    from utils import read_file
    
    df = read_file('data/all-scores.parquet')
    
    # Create crosstab of TEG vs Stableford points
    crosstab = pd.crosstab(df['TEGNum'], df['Stableford'], normalize='index') 
    
    # Round to 1 decimal place
    crosstab = crosstab.round(3)
    
    return crosstab
```

```python
def get_score_distribution_by_teg(score_field = 'Stableford'):
    """
    Returns a DataFrame showing the distribution of stableford points by TEG.
    
    Returns:
        pd.DataFrame: TEG as index, stableford points (0-max) as columns, 
                     cells show percentage of holes with that score
    """
    from utils import read_file
    
    df = read_file('data/all-scores.parquet')
    
    # Create crosstab of TEG vs Stableford points
    crosstab = pd.crosstab(df['TEGNum'], df[score_field], normalize='index') 
    
    # Round to 1 decimal place
    crosstab = crosstab.round(3)
    
    return crosstab
```
